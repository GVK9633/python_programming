```python
# Approach-1
ShippingDate =[]
ShippingMonth =[]
ShippingYear =[]

for x in df1['Ship Date']:
    result = str(x).split('-')
    ShippingDate.append(result[0])
    ShippingMonth.append(result[1])
    ShippingYear.append(result[2])
df1['Shipping Date'] = ShippingDate
df1['Shipping Month'] = ShippingMonth
df1['Shipping Year'] = ShippingYear

df1
   
    
```

```python
# approach-2
df1['Date'] = df1['Order Date'].apply(lambda x: x.split('-')[0])  # Extract the Month
df1['Month'] = df1['Order Date'].apply(lambda x: x.split('-')[1])   # Extract the Day
df1['Year'] = df1['Order Date'].apply(lambda x: x.split('-')[2])   # Extract the Year
df1
```
