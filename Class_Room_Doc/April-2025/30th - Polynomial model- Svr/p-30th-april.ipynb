```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset= pd.read_csv(r'./emp_sal.csv')
dataset
```

```python
x= dataset.iloc[:,1:2].values
y=dataset.iloc[:,2].values
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(x,y)
```

```python
plt.scatter(x,y,color='red')
plt.plot(x,lin_reg.predict(x),color='blue')
plt.title('Linear regression graph')
plt.xlabel('position level')
plt.ylabel('salary')
plt.show()
```

```python

# LinearRegression
lin_modle_pred = lin_reg.predict([[6.5]])
lin_modle_pred
```

```python
# polynomial degree 2
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures()
x_poly = poly_reg.fit_transform(x)
x_poly
```

```python
poly_reg.fit(x_poly,y)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(x_poly,y)
```

```python
plt.scatter(x,y,color='red')
plt.plot(x,lin_reg_2.predict(poly_reg.fit_transform(x)),color='blue')
plt.title('Truth or bluff (polynomial regression)')
plt.xlabel('position level')
plt.ylabel('salary')
plt.show()
```

```python
lin_modle_pred = lin_reg_2.predict(poly_reg.fit_transform([[6.5]]))
lin_modle_pred
```

```python
# SVR (support vector regression)
# Importing the SVR model
from sklearn.svm import SVR

# Creating the SVR model
regressor_svr = SVR(kernel='poly', degree=4, gamma='auto')

# Fitting the SVR model (Make sure 'x' and 'y' are properly defined datasets)
regressor_svr.fit(x, y)

# Predicting a value
y_pred_svr = regressor_svr.predict([[6.5]])

# Printing the predicted value
print(y_pred_svr)
```

```python
#1-may-20225
from sklearn.neighbors import KNeighborsRegressor
regressor_knn= KNeighborsRegressor()
regressor_knn.fit(x,y)
y_pred_kkn = regressor_knn.predict([[6.5]])
print(y_pred_kkn)
```

```python
#usecase-2
from sklearn.neighbors import KNeighborsRegressor
regressor_knn= KNeighborsRegressor(n_neighbors=4,weights='distance')
regressor_knn.fit(x,y)
y_pred_kkn = regressor_knn.predict([[6.5]])
print(y_pred_kkn)
```

```python
# cart or tree algo
from sklearn.tree import DecisionTreeRegressor
regressor_dtr = DecisionTreeRegressor()
regressor_dtr.fit(x,y)
y_pred_dtr = regressor_dtr.predict([[6.5]])
print(y_pred_dtr)
```

```python
# cart or tree algo hyper parameter
from sklearn.tree import DecisionTreeRegressor
regressor_dtr = DecisionTreeRegressor(criterion='absolute_error',splitter='random',random_state=0)
regressor_dtr.fit(x,y)
y_pred_dtr = regressor_dtr.predict([[6.5]])
print(y_pred_dtr)
```

```python
#Random forest algo
from sklearn.ensemble import RandomForestRegressor
regressor_rf = RandomForestRegressor()
regressor_rf.fit(x,y)
y_pred_rf = regressor_rf.predict([[6.5]])
print(y_pred_rf)
```

```python
# use case-1 Random forest algo
from sklearn.ensemble import RandomForestRegressor
regressor_rf = RandomForestRegressor(random_state=0)
regressor_rf.fit(x,y)
y_pred_rf = regressor_rf.predict([[6.5]])
print(y_pred_rf)
```

```python
# use case-2 Random forest algo
from sklearn.ensemble import RandomForestRegressor
regressor_rf = RandomForestRegressor(random_state=0,n_estimators=20)
regressor_rf.fit(x,y)
y_pred_rf = regressor_rf.predict([[6.5]])
print(y_pred_rf)
```
